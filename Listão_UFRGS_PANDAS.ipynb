# Buscando os meus possíveis colegas no Listão :)

##### O listão online da UFRGS é uma série de listas em ordem alfabética. Irei importar para o Notebook e descobrir meus possíveis colegas de faculdade!  

##### Link de acesso dos dados: https://vestibular.ufrgs.br/cv2022/listao/

###Importando as bibliotecas

```python
import pandas as pd
```

### Pegando a lista de selecionados com o Pandas

```python
letras = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'r', 's', 't', 'u', 'v', 'w', 'y', 'z']  # Não há candidatos nas letras "q" e "x".
```

```python
listao = []

for letra in letras:
  listao.append((pd.read_html('https://vestibular.ufrgs.br/cv2022/listao/arquivo_'+letra+'.html')))
```

```python
# Visualizando...

listao[0][0].head()
```

### Concatenando as listas de DataFrames em um só.

```python
novo_listao = pd.concat([listao[0][0]])

for num in range(1, 24):
  novo_listao = novo_listao.append(pd.concat([listao[num][0]]))
```

```python
novo_listao.sample(5)
```

# Pegando os aprovados em Ciência da Computação

```python
# É possível ver qualquer outro curso apenas trocando o nome da variável curso.

curso = ['Ciência da Computação']
selecao = novo_listao['Curso'].isin(curso)
dados = novo_listao[selecao].reset_index()
del dados['index']
```

```python
dados.head(5)
```

###Fazendo algumas análises

```python
# Quem vai entrar no primeiro semestre?

dados.query('Semestre == 1')
```

```python
# Quem vai entrar no Segundo Semestre?

dados.query('Semestre == 2')
```

```python
# Tirando o caractere "°"
dados['Classificação'].replace(to_replace="°", value="", inplace=True, regex = True)

# Transformando para inteiro
dados['Classificação'] = dados['Classificação'].astype(int)

# Ordenando a Classificação:
dados_ordenados = dados.sort_values(by = "Classificação").reset_index()

del dados_ordenados['index']
```

```python
dados_ordenados.head(5)
```

```python
# primeiro semestre ordenado

primeiro_semestre = dados_ordenados.query('Semestre == 1').reset_index()
del primeiro_semestre['index']
primeiro_semestre
```

```python
# segundo semestre ordenado

segundo_semestre = dados_ordenados.query('Semestre == 2').reset_index()
del segundo_semestre['index']
segundo_semestre
```

# Transformando para um arquivo .csv



```python
# O dataframe do curso de Ciência da Computação em ordem de classificação:
dados_ordenados.to_csv('listao_ufrgs_cc.csv', index = False)

# O dataframe do listao em ordem alfabética:
novo_listao.to_csv('listao_ufrgs.csv', index = False)
```
